```python
from fourier import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
def graph_series(f, terms):
    series = FourierSeries(f, terms)

    x = np.linspace(-math.pi, math.pi, 1000)
    s = np.vectorize(series.eval)

    plt.grid()

    plt.plot(x, f(x), color = 'red')
    plt.plot(x, s(x), color = 'blue')
    plt.show()
```

```python
graph_series(lambda x: x, 5)
```

```python
f = lambda x : -1 if(x < 1) else 1
graph_series(np.vectorize(f), 10)
```

```python
f = lambda x : x**2 + x**3
graph_series(f, 15)
```

```python
f = lambda t : math.cos(math.exp(t))
graph_series(np.vectorize(f), 30)
```
